T1. On considère l'expérience aléatoire qui consiste à tirer uniformément et indépendamment deux entiers sur $[\![ 1, n ]\!]$, $n \in \mathbb{N}$.

Il faut montrer que
$$
\mathcal{P}_n:=\mathbb{P}[P_n]=\frac{6}{\pi^2}+O\left(\frac{\ln n}{n}\right).
$$

a) L'événement totale $\Omega = \{\text{à tirer uniformément et indépendamment deux entiers}\}$ et on a
$$
    |\Omega| = \frac{ |[\![1,n]\!] \times [\![1,n]\!]|}{2!}=\frac{n^2}{2}
$$
parce que les deux entiers ne sont pas ordonnée. Également, notons que deux entiers $a$ et $b$ sont premiers entre eux ssi $PGCD(a,b)=PGCD(b,a)=1$ et donc $(a,b) \in A_n$ et $(b,a) \in A_n$. En revance, si $(a,b) \in A_n$, on a $(b,a) \in A_n$ et alors $a$ et $b$ sont entiers entre eux. C'est-à-dire,
$$
|A_n| = 2|P_n|
$$
et alors
$$
\mathcal{P}_n = \frac{|P_n|}{|\Omega_n|} = \frac{|A_n|}{n^2}.
$$
b) Soit $(a,b) \not\in A_n$. C'est-à-dire, $PGCD(a,b) >1$. Soit $p_i\mid PGCD(a,b)$, donc $p_i \mid a$ et $p_i \mid b$ et donc $(a,b) \in D_i \subset \bigcup_{i=1}^{\pi_n}D_i$.

En revanche, si $(a,b) \in \bigcup_{i=1}^{\pi_n}D_i$, il existe $i \in [\![1,\pi(n)]\!]$ tel que $(a,b) \in D_i$, donc $p_i \mid a$ et $p_i \mid b$. À ce moment-là, on a $p_i|PGCD(a,b)$ et donc $(a,b) \not\in A_n$. Du coup on a montré que
$$
A_n^c = \bigcup_{i=1}^{\pi_n}D_i
$$
qui est equivalent que ce que on veut montrer.
c) Par (b), on a
$$
\begin{aligned}
    |A_n| &=  \left|[\![1,n]\!] \times [\![1,n]\!]\right| - \left|\left(\bigcup_{i=1}^{\pi_n}D_i\right)\right| \\
           &= n^2 - \sum_{I \subset [\![1,\pi(n)]\!],\, I \ne \varnothing}(-1)^{|I|+1}\left|\bigcap_{i \in I}D_i\right| \\
           &= n^2 + \sum_{I \subset [\![1,\pi(n)]\!],\, I \ne \varnothing}(-1)^{|I|}\left|\bigcap_{i \in I}D_i\right|
\end{aligned}
$$
Pour $I \subset [\![1,\pi(n)]\!],\, I \ne \varnothing$, si on met $\ell = \prod_{i \in I}p_i$, on a $(-1)^{|I|}=\mu(\ell)$. En revanche, notons que $(a,b) \in \bigcap_{i \in I}D_i$ ssi $\ell|a$ et $\ell | b$. C'est-à-dire $a,b \in \{\ell,2\ell,\dots,\lfloor \frac{n}{\ell}\rfloor\ell\}$ et donc
$$
    |\bigcap_{i \in I}D_i| = |\{\ell,2\ell,\dots,\lfloor \frac{n}{\ell}\rfloor\ell\} \times \{\ell,2\ell,\dots,\lfloor \frac{n}{\ell}\rfloor\ell\}| = \left\lfloor \frac{n}{\ell} \right\rfloor^2
$$
Après, notons que pour $\ell \in [\![1,n]\!]$, $\mu(\ell)\ne 0$ ssi $\ell=\prod_{i \in I}p_i$ où $I \subset [\![1,\pi(n)]\!],\, I \ne \varnothing$ ou $\ell=1$. Donc
$$
    \sum_{\ell=1}^{n}\mu(\ell)\left\lfloor \frac{n}{\ell} \right\rfloor^2 = \sum_{\ell \in [\![1,n]\!]}^{\mu(\ell) \ne 0}\mu(\ell)\left\lfloor \frac{n}{\ell} \right\rfloor^2 = \sum_{I \subset [\![1,\pi(n)]\!],\, I \ne \varnothing}(-1)^{|I|}\left|\bigcap_{i \in I}D_i\right|+\mu(1)n^2 = |A_n|
$$

d) **Lemme.** Pour $s>1$, on a
$$
\sum_{n=1}^{\infty}\frac{\mu(n)}{n^s}=\frac{1}{\zeta(s)}.
$$
*Preuve.* Notons que
$$
    \begin{aligned}
        \frac{1}{\zeta(s)} &= \prod_{i=1}^{\infty}\left(1-\frac{1}{p_i^s}\right) \\
                           &= 1  -\sum_{i}\frac{1}{p_i^s} + \sum_{i<j}\frac{1}{p_i^s p_j^s} - \sum_{i<j<k}\frac{1}{p_i^s p_j^s p_k^s} + \\
                           &= \sum_{n=1}^{\infty}\frac{\mu(n)}{n^s}
    \end{aligned}
$$
$\square$

Pour la question (d), soit $\{x\}=x-\lfloor x \rfloor$. Par exemple, $\{2.3\}=0.3$. Alors
$$
    \begin{aligned}
        \mathcal{P}_n = \frac{1}{n^2}|A_n| &= \sum_{\ell=1}^{n}\frac{1}{n^2}\mu(\ell)\left\lfloor \frac{n}{\ell} \right\rfloor^2 \\
            &= \frac{1}{n^2}\sum_{\ell=1}^{n}\mu(\ell)\left(\frac{n}{\ell}-\left\{\frac{n}{\ell}\right\}\right)^2 \\
            &= \sum_{\ell=1}^{n}\mu(\ell)\frac{1}{\ell^2} -\frac{2}{n^2}\sum_{\ell=1}^{n}\mu(\ell)\left(\frac{n}{\ell}\right)\left\{\frac{n}{\ell}\right\} + \frac{1}{n^2}\sum_{\ell=1}^{n}\mu(\ell)\left\{\frac{n}{\ell}\right\}^2
    \end{aligned}
$$
Et donc
$$
    \begin{aligned}
        \mathcal{P}_n &= \left|\sum_{\ell=1}^{n}\mu(\ell)\frac{1}{\ell^2} -\frac{2}{n^2}\sum_{\ell=1}^{n}\mu(\ell)\left(\frac{n}{\ell}\right)\left\{\frac{n}{\ell}\right\} + \frac{1}{n^2}\sum_{\ell=1}^{n}\mu(\ell)\left\{\frac{n}{\ell}\right\}^2\right| \\
                     &\leq \left|\sum_{\ell=1}^{n}\mu(\ell)\frac{1}{\ell^2}\right| + \left|\frac{2}{n^2}\sum_{\ell=1}^{n}\mu(\ell)\left(\frac{n}{\ell}\right)\left\{\frac{n}{\ell}\right\}\right| + \left|\frac{1}{n^2}\sum_{\ell=1}^{n}\mu(\ell)\left\{\frac{n}{\ell}\right\}^2\right|
    \end{aligned}
$$
On utilise le fait que $|\mu(\ell)| \le 1$ et $\left\{\frac{n}{\ell}\right\} \le 1$ pour obtenir que
$$
    \begin{aligned}
        \mathcal{P}_n &\leq \left|\frac{1}{\zeta(2)} - \sum_{\ell=n+1}^{\infty}\frac{1}{\ell^2}\right|+\left|\frac{2}{n}\sum_{\ell=1}^{n}\frac{1}{\ell}\right| + \left|\sum_{\ell=1}^{n}\frac{1}{n^2} \right| \\
                      &\leq \frac{6}{\pi^2} + \sum_{\ell=n+1}^{\infty}\frac{1}{\ell^2} + \frac{2}{n}\sum_{\ell=1}^{n}\frac{1}{\ell} + \sum_{\ell=1}^{n}\frac{1}{n^2}
    \end{aligned}
$$
Parce que $\sum_{\ell=n+1}^{\infty}\frac{1}{\ell^2} \sim \frac{1}{n}=O\left(\frac{\ln n}{n}\right)$ (considérons l'intégrale $\int_{n+1}^{\infty}\frac{1}{x^2}\mathrm{d}x$), $\frac{2}{n}\sum_{\ell=1}^{n}\frac{1}{\ell} \sim \frac{\ln n}{n}$ (considérons $\lim_{n\to\infty}\left(1+\frac{1}{2}+\dots+\frac{1}{n}-\ln{n}\right)=\gamma<\infty$) et $\sum_{\ell=1}^{n}\frac{1}{n^2}=\frac{1}{n}$, on obtient $\sum_{\ell=n+1}^{\infty}\frac{1}{\ell^2} + \frac{2}{n}\sum_{\ell=1}^{n}\frac{1}{\ell} + \sum_{\ell=1}^{n}\frac{1}{n^2}=O\left(\frac{\ln n}{n}\right)$ et donc
$$
    \mathcal{P}_n = \frac{6}{\pi^2}+O\left(\frac{\ln n}{n}\right)
$$

Dans un premier temps, nous nous concentrons sur la méthode d'inversion qui permet d'obtenir des réalisations uniformes indépendantes sur $\![ 0, 1]\!$ à l'aide de réalisations uniformes indépendantes sur $[\![ 1, M ]\!]$.
En effet, il s'agira d'écrire la fonction qui retourne 1 si x est égale à 0, la partie entière de $nx$ si $nx$ $\in \mathbb{N}$ et la partie entière de $nx$ plus 1 sinon, avec $x \in [ 0, 1 ]$.

```python
# Des bibliothèques classiques de Python
import numpy as np
import scipy as scp
from scipy.integrate import quad
import scipy.stats as sps
import matplotlib.pyplot as plt
import random
import time
import tracemalloc
```

```python
def realisation_uniforme(n):
    u=random.random()
    #u=random.randint(1,n)
    if (u==0):
        return 1
    elif(n*u-np.floor(n*u)==0):
         return n*u
    else:
         return np.floor(n*u)+1
    return u
```

Maintenant, nous comparons l'éfficacité de cette méthode à celle de la fonction randint()

```python
randUniform = []
randInt = []
tracemalloc.start()

start_time = time.time()
for i in range(10000):
    randUniform.append(realisation_uniforme(10000))
end_time = time.time()

current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()
time_uniforme = (end_time - start_time) * 1000
memory_uniforme = peak / 1024

tracemalloc.start()
start_time = time.time()
for i in range(10000):
    randInt.append(random.randint(0,10000))
end_time = time.time()
tracemalloc.stop()
time_randint = (end_time - start_time) * 1000
memory_randint = peak / 1024

time_comparison = [time_uniforme,time_randint]
memory_comparison = [memory_uniforme,memory_randint]
label = ['Uniforme','randint']
plt.figure(figsize=(14,14))
plt.subplot(2,2,1)
plt.plot(range(10000),randUniform,'+',color='b')
plt.title("Utilisant realisation_uniforme")
plt.subplot(2,2,2)
plt.plot(range(10000),randInt,'+',color='r')
plt.title("Utilisant randint")
plt.subplot(2, 2, 3)
plt.ylabel('Time')
plt.title('Time Comparison')
plt.bar(label,time_comparison,color=['b','r'])
plt.xlabel("Functions")
plt.ylabel("Time (ms)")
plt.subplot(2,2,4)
plt.ylabel("Memory")
plt.title("Peak memory comparison")
plt.bar(label,memory_comparison,color=['b','r'])
plt.xlabel("Functions")
plt.ylabel("Memory (KB)")
plt.show()
```

D'après les résultats du temps d'exécution des deux méthodes, on conclut que la fonction randint() est la plus éfficace.

Considérons la variable de Bernoulli $B_n$ qui renvoie 1 si le tirage des deux entiers compris entre 1 et $n$ ci-dessus donne des entiers premiers entre eux et 0 sinon. 

```python
def realisation_Bi(n):
   #   u1=random.randint(1,n)
   #   u2=random.randint(1,n)
     sample = random.sample(range(n+1),2)
     u1 = sample[0]
     u2 = sample[1]
     if(np.gcd(u1,u2)==1):
        return 1
     else:
        return 0
```

Considérons également $\mathcal{P}^M_n:=\frac{\sum_{i=1}^n B_n^i}{M}$, la moyenne empirique des $B_n^i$ suivant tous la même loi que $B_n$.

```python
def probabilité(n,M):
    B=0
    for i in range(1,M+1):
        B=B+(realisation_Bi(n))
    return B/float(M)
```

${P}^M_n$ est un estimateur de ${P}_n$ qui d'après la loi des grands nombres, converge presque sûrement vers ${P}_n$ quand M tends vers l'infini. Mais puisque d'après ce qu'on a fait précédemment ${P}_n$ tends vers $\frac{6}{\pi^2}$, on devrait également voir que ${P}^M_n$ tend vers $\frac{6}{\pi^2}$ quand n et M tendent vers l'infini.


```python

def affichage_proba(n,M):
    x=range(1,n+1)
    y=[]
    for i in x:
        y.append(probabilité(i,M))
    plt.title('Représentation de $\\mathcal{P}^M_n$ en fonction de $n$')
    plt.plot(x, y,label='La valeur donnée par l\'éstimateur')
    plt.axhline(y = 6/np.pi**2,linestyle='--',color='r',label='Valeur théorique $\\mathcal{P}_n=\\frac{6}{\\pi^2}$')
    plt.xlabel('n')
    plt.ylabel('$\\mathcal{P}^M_n$')
    plt.legend()
    plt.show()
    
affichage_proba(200,3000)
```

T6. Parce que $\mathcal{B}_i$ sont indépendantes, on a immédiatement
$$
m(n) = \mathbb{E}(\Pi(n)) = \sum_{k=2}^{n}\mathbb{E}(\mathcal{B}_i) = \sum_{k=2}^{n}\frac{1}{\log k}
$$
et également
$$
\mathbb{V}(\Pi(n)) = \sum_{k=2}^{n}\mathbb{V}(\mathcal{B}_i) = \sum_{k=2}^{n}\frac{1}{\log k}\left(1-\frac{1}{\log k}\right)
$$
Par la définition de variance, on a
$$
\begin{aligned}
\mathbb{V}(\Pi(n)) &= \sum_{k=2}^{n}\frac{1}{\log k}\left(1-\frac{1}{\log k}\right) \\
                   &= \mathbb{E}(|\Pi(n)-m(n)|^2) \\
                   &\ge \mathbb{E}(|\Pi(n)-m(n)|^2 \chi_{\{|\Pi(n)-m(n)|^2 \ge n^{4/3}\}}) \\
                   &\ge n^{4/3}\mathbb{P}(|\Pi(n)-m(n)|^2 \ge n^{4/3})
\end{aligned}
$$

Pour majorer la variance, notons que $\int_2^x \frac{1}{\log x}\left(1-\frac{1}{\log x}\right)dx = \frac{x}{\log x} - \frac{2}{\log 2}$ et donc $\sum_{k=2}^{n}\frac{1}{\log k}\left(1-\frac{1}{\log k}\right) \sim \frac{n}{\log n}$ ou autrement dit, $\sum_{k=2}^{n}\frac{1}{\log k}\left(1-\frac{1}{\log k}\right) \le C \frac{n}{\log n}$ pour certain $C>0$.

Et nous déduisons que $\mathbb{P}(|\Pi(n)-m(n)|^2 \ge n^{4/3}) = \mathbb{P}(|\Pi(n)-m(n)| \ge n^{2/3}) \le C\frac{1}{n^{1/3}\log n}$. Donc
$$
0 \le \lim_{n\to\infty}\mathbb{P}(|\Pi(n)-m(n)|^2 \ge n^{4/3}) \le \lim_{n\to\infty}C\frac{1}{n^{1/3}\log n} = 0
$$
et donc
$$
\lim_{n\to\infty}\mathbb{P}(|\Pi(n)-m(n)| < n^{2/3}) = \lim_{n\to\infty} [1 - \mathbb{P}(|\Pi(n)-m(n)| \ge n^{2/3})] = 1
$$

Pour $A_n = \{|\Pi(n^4)-m(n^4)| \ge n^{8/3}\}$, par l'inegalité que l'on vient d'obtenir,
$$
\mathbb{P}(A_n) \le C n^{-1/3}\log(n^4)^{-1} = C' n^{-4/3} \log(n)^{-1};\; C,C'>0.
$$
mais la série $\sum_{n=2}^{\infty}\frac{1}{n^{4/3}\log n}$ converge parce que $0 < \frac{1}{n^{4/3}\log n} < \frac{1}{n^{4/3}}$ et $\sum\frac{1}{n^{4/3}}$ converge. Alors $\sum_{n=2}^{\infty}\mathbb{P}(A_n)<\infty$ et puis $\mathbb{P}(\liminf_{n\to\infty} A_n^c)=1$ (Borel-Cantelli).

Pour $B_k=\{|\Pi(k)-m(k)| \le Ck^{3/4}\}$, notons que 
$$
B_{n^4} = \{|\Pi(n^4)-m(n^4)| \le C n^3\} \supset A_n^c
$$
et de même lorsque $k \in [\![n^4,(n+1)^4]\!]$, on a
$$
|\Pi(k)-m(k)| = |\Pi(k)-\Pi(n^4)|+|\Pi(n^4)-m(n^4)|+|m(k)-m(n^4)|
$$

Parce qu'il y a $O(n^3)$ entiers entre $k$ et $n^4$, on voit que $|\Pi(k)-\Pi(n^4)|=O(n^3)$ et $|m(k)-m(n^4)|=O(n^3)$. Lorsque $B_{n^4}$ est vrai, on sait que $|\Pi(n^4)-m(n^4)|\le Cn^3$ et donc $|\Pi(k)-m(k)|=O(n^3)$ aussi. Donc dans tous les cas, $B_k \supset A_n^c$ pour $k \in [\![n^4,(n+1)^4]\!]$ et donc $\bigcap_{m=n}^{\infty}B_m \supset \bigcap_{m=n}^{\infty}A_m^c$. Cela signifie que
$$
\liminf_{k}B_k \supset \liminf_{n}A_n^c \implies \mathbb{P}[\liminf_{k}B_k] = 1
$$

T7. Rappelons le TCL de Lyapunov
> **Théorème - Théorème central limite (Lyapunov).** Soit $(X_1,\dots,X_n)$ une suite de variables aléatoires tel que $\mu_i=\mathbb{E}[X_i]<\infty$ et $\sigma_i^2 = \mathbb{E}[X_i]$. Définissons
>$$
>s_n^2 = \sum_{i=1}^{n} \sigma_i^2.
>$$
>S'il existe $\delta > 0$ tel que
>$$
>\lim_{n \to \infty} \frac{1}{s_n^{2+\delta}}\sum_{i=1}^{n}\mathbb{E}[X_i-\mu_i|^{2.+\delta}] = 0,
>$$
>alors
>$$
>\frac{1}{s_n} \sum_{i=1}^{n}(X_i - \mu_i) \xrightarrow{\mathcal{L}}\mathcal{N}(0,1).
>$$

Pour utiliser ce théorème, en définissant $X_n = \mathcal{B}_n$ et $X_0 = X_1 = 0$, il faut noter que $|X_n|$ est bornée par $1$ uniformément et donc
$$
\sum_{k=2}^{n}\frac{1}{s_n^3}\mathbb{E}[|X_k|^3] \le \sum_{k=2}^{n}\frac{1}{s_n^3}\mathbb{E}[|X_k|^2]=\frac{1}{s_n} \to 0\;(n\to\infty)
$$

Référence : Billingsley, Patrick (1995). Probability and Measure (3rd ed.). John Wiley & Sons.

T8. Nous allons utiliser le théorème de Kolmogorov (1929). [Kai Lai Chung, *A Course in Probability Theory*]
> **Théorème.** Soit $\{X_n\}$ une suite de v.a. tel que $\mathbb{E}(X_n)=0$. Soit $S_n=\sum_{j=1}^{n}X_j$. Supposons que pour $s_n^2=\mathbb{V}(S_n)$, on a 
>$$
>\sup_\omega |X_n(\omega)| = o \left(\frac{s_n}{\sqrt{\log\log s_n}}\right),
>$$
>on aura que, presque sûrement,
>$$
>\limsup_{n\to\infty}\frac{S_n}{\sqrt{2s_n^2\log\log s_n}}=1.
>$$

Dans notre problème, on met $X_n = \mathcal{B}_n - \mathbb{E}(\mathcal{B}_n)$ pour $n \ge 2$ et $X_1 = 0$. On a $\sup|X_n| \le 1 = o \left(\frac{s_n}{\sqrt{\log\log s_n}}\right)$. On a immédiatement
$$
\limsup_{n\to\infty}\frac{\Pi(n)-m(n)}{\sqrt{2\mathbb{V}(\Pi(n))\log\log(\sqrt{\mathbb{V}(\Pi(n))})}}=1
$$
Rapellons que $\mathbb{V}(\Pi(n))= \mathbb{V}(S_n) = \frac{n}{\log n}+O(1)$ et donc
$$
\begin{aligned}
    2\mathbb{V}(\Pi(n))\log\log(\sqrt{\mathbb{V}(\Pi(n))}) &= 2\left(\frac{n}{\log n}+O(1)\right)\log\log\sqrt{\frac{n}{\log n}+O(1)}.
\end{aligned}
$$
Notons que
$$
\begin{aligned}
    \lim_{n\to\infty}\frac{\log\log\sqrt{\frac{n}{\log n}+O(1)}}{\log\log(n)} &= \lim_{n\to\infty}\frac{\log\log\sqrt{\frac{n}{\log{n}}}}{\log\log n} \\
                                                                                    &= \lim_{n\to\infty}\frac{\log\left(\frac{1}{2}\log\frac{n}{\log{n}}\right)}{\log\log n} \\
                                                                                    &= \lim_{n\to\infty}\frac{-\log 2 + \log\log\frac{n}{\log{n}}}{\log\log n} \\
                                                                                    &= \lim_{n\to\infty}\frac{-\log 2 + \log\log n - \log\log\log n}{\log\log n} \\
                                                                                    &= \lim_{n\to\infty}\frac{-\log 2}{\log\log n} + \lim_{n\to\infty}\frac{\log\log n}{\log\log n} + \lim_{n\to\infty}\frac{\log\log\log n}{\log\log n} \\
                                                                                    &= 1
\end{aligned}
$$
et donc
$$
\lim_{n\to\infty}\frac{\sqrt{2\mathbb{V}(\Pi(n))\log\log\sqrt{\mathbb{V}(\Pi(n))}}}{\sqrt{2\left(\frac{n}{\log n}\right)\log\log n}} = \lim_{n\to\infty}\frac{\sqrt{2\left(\frac{n}{\log n} + O(1)\right)\log\log\sqrt{\frac{n}{\log n}+O(1)}}}{\sqrt{2\left(\frac{n}{\log n}\right)\log\log n}} = 1.
$$

Finalement,
$$
\begin{aligned}
\limsup_{n\to\infty}\frac{\Pi(n)-\mathrm{Li(n)}}{\sqrt{2\frac{n}{\log n}\log\log n}} &= \limsup_{n\to\infty}\frac{\Pi(n)-m(n)+m(n)-\mathrm{Li}(n)}{\sqrt{2\frac{n}{\log n}\log\log n}} \\
                                                                                     &= \limsup_{n\to\infty}\frac{\Pi(n)-m(n)+O(1)}{\sqrt{2\mathbb{V}(\Pi(n))\log\log\sqrt{\mathbb{V}(\Pi(n))}}} \\
                                                                                     &= 1
\end{aligned}
$$

T9. Posons $E_m^\eta = \bigcap_{i=1}^{\lceil \eta \ln(m)^2 \rceil}\{\mathcal{B_{m+i}}=0\}$. Soit $M=\lceil \eta \ln(m)^2 \rceil$. Alors par définition, on a 
$$
\mathbb{P}[E_m^\eta] = \prod_{k=1}^{M}\left(1-\frac{1}{\log (k+m)}\right)
$$

Alors a) est facile parce que

$$
\mathfrak{p}_{n+1} - \mathfrak{p}_n > \lceil\eta \log (\mathfrak{p})^2\rceil \ge \eta \log (\mathfrak{p})^2 \text{ pour un infinité de $n$ } \iff \text{ il y a un infinité de $n$ dans $E_m^\eta$. } 
$$
et c'est équivalent à dire
$$
\mathbb{P}(\limsup\{\mathfrak{p}_{n+1}-\mathfrak{p}_n\}> \eta \log (\mathfrak{p}_n)^2) = \mathbb{P}(\limsup_m E_m^\eta)
$$

Pour b), rappelons que $\log(1+x)=x+O(x^2)$ et alors $1+x = \exp(x+O(x^2))$, et la question est équivalent à montrer que $\mathbb{P}(E_m^\eta)=\frac{O(1)}{m^\eta}$ Donc
$$
\begin{aligned}
\mathbb{P}[E_m^\eta] &= \prod_{k=1}^{M}\left(1-\frac{1}{\log (k+m)}\right) \\
                     &= \prod_{k=1}^{M}\exp\left(-\frac{1}{\log (k+m)}+O\left(\frac{1}{\log (k+m)}\right)^2\right) \\
                     &= \exp\left(\sum_{k=1}^{M}\frac{-1}{\log (k+m)}+O\left(\sum_{k=1}^{M}\left(\frac{1}{\log (k+m)}\right)^2\right)\right)
\end{aligned}
$$
Notons que 
$$
\begin{aligned}
\sum_{k=1}^{M}\left(\frac{1}{\log (k+m)}\right)^2 &< \sum_{k=1}^{M}\left(\frac{1}{\log (m)}\right)^2 \\
                                                         &= \frac{\lceil \eta \log(m)^2\rceil}{\log(m)^2}
\end{aligned}
$$
est bornée par un nombre determiné par $\eta$. Donc $O\left(\sum_{k=1}^{M}\left(\frac{1}{\log (k+m)}\right)^2\right)=O(1)$. 

Dans le même temps, 

$$
\sum_{k=1}^{M}\frac{1}{\log(m+k)} = \frac{M}{\log(m+M+1)}-\sum_{k=1}^{M}k\left(\frac{1}{\log(m+k+1)}-\frac{1}{\log(m+k)}\right)=\eta\log(m)+O(1)
$$
Combinons tous les résultats qu'on vient d'obtenir, on a
$$
\mathbb{P}(E_m^\eta) = \frac{O(1)}{m^\eta}.
$$

Pour c), notons que lorsque $\eta>1$, on a
$$
\sum_{m=1}^{\infty}\mathbb{P}[E_m^\eta] \le \sum_{m=1}^{\infty}\frac{C_2}{m^\eta} < \infty
$$
et alors par Borel-Cantelli, 
$$
\mathbb{P}(\limsup\{\mathfrak{p}_{n+1}-\mathfrak{p}_n\}> \eta \log (\mathfrak{p}_n)^2) = \mathbb{P}(\limsup_m E_m^\eta) = 0
$$

En revanche, lorsque $0 < \eta < 1$, posons $m_1 = 2$ et $m_{k+1} = m_k + \lceil \eta \log m_k \rceil$. Par le raisonment par récurrence, on a que $m_k < e^5k(\log k)^2$ pour tous les $k=1,2,\dots$. Dans ce cas,
$$
\sum_{k=1}^{\infty}\mathbb{P}[E_{m_k}^\eta] \ge  \sum_{k=1}^{\infty}\frac{C_1}{m_k^\eta} \ge \sum_{k=1}^{\infty}\frac{C_1}{(e^5 k (\log k)^2)^\eta}=\infty
$$

Parce que $(E_{m_k})_k$ sont indépendants, on vois que par Borel-Cantelli, $\mathbb{P}(\limsup_m E_m^\eta) = \mathbb{P}(\limsup_k E_{m_k}^\eta)=1$.

Donc il y a un infinité de $n$ tel que $\mathfrak{p}_{n+1} - \mathfrak{p} > \eta \log(\mathfrak{p})^2$ pour tout $0<\eta < 1$. Donc
$$
\limsup_{n}\frac{\mathfrak{p}_{n+1}-\mathfrak{p}_n}{\log(\mathfrak{p_n})^2}=1
$$
presque sûrement.
